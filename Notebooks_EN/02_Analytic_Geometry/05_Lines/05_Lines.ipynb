# Solutions to Tasks Related to Equations of Lines on a Plane

---

## **1. The line passes through points $A(1, 2)$ and $B(3, 4)$. Find the equation of the line.**

### Solution:
1. Find the slope:
   $$ m = \frac{y_2 - y_1}{x_2 - x_1} = \frac{4 - 2}{3 - 1} = 1 $$
2. Use the point-slope form:
   $$ y - y_1 = m(x - x_1) $$
   Substituting $m = 1$, $x_1 = 1$, and $y_1 = 2$:
   $$ y - 2 = 1(x - 1) \implies y = x + 1 $$

**Equation:**  
$ y = x + 1 $

---

## **2. The line passes through point $A(1, 2)$ and is parallel to the line $y = 2x + 3$. Find the equation of the line.**

### Solution:
1. The slope of $y = 2x + 3$ is $m = 2$ (parallel lines have the same slope).  
2. Use the point-slope form:
   $$ y - y_1 = m(x - x_1) $$
   Substituting $m = 2$, $x_1 = 1$, and $y_1 = 2$:
   $$ y - 2 = 2(x - 1) \implies y = 2x $$

**Equation:**  
$ y = 2x $

---

## **3. The line passes through point $A(1, 2)$ and is perpendicular to the line $y = 2x + 3$. Find the equation of the line.**

### Solution:
1. The slope of the given line is $m = 2$.  
   For perpendicular lines, the slope is:
   $$ m_\perp = -\frac{1}{m} = -\frac{1}{2} $$
2. Use the point-slope form:
   $$ y - y_1 = m_\perp(x - x_1) $$
   Substituting $m_\perp = -\frac{1}{2}$, $x_1 = 1$, and $y_1 = 2$:
   $$ y - 2 = -\frac{1}{2}(x - 1) \implies y = -\frac{1}{2}x + \frac{5}{2} $$

**Equation:**  
$ y = -\frac{1}{2}x + \frac{5}{2} $

---

## **4. Intersection point of $y = 2x + 3$ and $y = 3x + 2$ and angle between them.**

### Solution:
1. **Find the intersection point:**
   Solve:
   $$ 2x + 3 = 3x + 2 \implies x = 1, \, y = 5 $$
   Intersection point: $(1, 5)$.

2. **Calculate the angle:**
   The slopes are $m_1 = 2$ and $m_2 = 3$.  
   Use the formula:
   $$ \tan \theta = \left| \frac{m_1 - m_2}{1 + m_1m_2} \right| = \left| \frac{2 - 3}{1 + 2 \cdot 3} \right| = \frac{1}{7} $$
   Angle:
   $$ \theta = \arctan\left(\frac{1}{7}\right) $$

---

## **5. Line passing through $A(1, 2)$ and parallel to vector $\vec{v} = [2, 3]$.**

### Solution:
1. The slope is given by:
   $$ m = \frac{3}{2} $$
2. Use the point-slope form:
   $$ y - y_1 = m(x - x_1) $$
   Substituting $m = \frac{3}{2}$, $x_1 = 1$, and $y_1 = 2$:
   $$ y - 2 = \frac{3}{2}(x - 1) \implies y = \frac{3}{2}x + \frac{1}{2} $$

**Equation:**  
$ y = \frac{3}{2}x + \frac{1}{2} $

---

## **6. Example of a line perpendicular and parallel to $y = 2x + 3$.**

### Solution:
1. **Parallel line:**  
   Slope: $m = 2$. Example:  
   $$ y = 2x + 1 $$

2. **Perpendicular line:**  
   Slope: $m = -\frac{1}{2}$. Example:  
   $$ y = -\frac{1}{2}x + 4 $$

---

## **7. Distance from point $A(1, 2)$ to the line $y = 2x + 3$.**

### Solution:
1. Rewrite the line in general form:
   $$ -2x + y - 3 = 0 $$
   Here, $a = -2$, $b = 1$, and $c = -3$.

2. Use the distance formula:
   $$ d = \frac{|ax_1 + by_1 + c|}{\sqrt{a^2 + b^2}} $$
   Substituting $x_1 = 1$, $y_1 = 2$:
   $$ d = \frac{|-2(1) + 1(2) - 3|}{\sqrt{(-2)^2 + 1^2}} = \frac{3}{\sqrt{5}} $$

---

## **8. Line intersecting axes at $A(2, 0)$ and $B(0, 3)$.**

### Solution:
1. Find the slope:
   $$ m = \frac{3 - 0}{0 - 2} = -\frac{3}{2} $$
2. Use the point-slope form:
   $$ y - 0 = -\frac{3}{2}(x - 2) \implies y = -\frac{3}{2}x + 3 $$

**Equation:**  
$ y = -\frac{3}{2}x + 3 $

---

## **9. Angle between the line $y = x + 3$ and the $Ox$ axis.**

### Solution:
1. The slope of the line is $m = 1$.  
2. The angle is:
   $$ \theta = \arctan(m) = \arctan(1) = 45^\circ $$

---

## **10. Vector perpendicular to the line $x + y + 1 = 0$.**

### Solution:
1. Rewrite the line in vector form:  
   Direction vector: $[1, 1]$.
2. A perpendicular vector is:
   $$ \vec{v_\perp} = [-1, 1] $$
