# Conic Sections: Comprehensive Explanation and Concepts

---

## **1. Circle**

### **Definition:**
A circle is the set of all points in a plane that are equidistant from a fixed point called the center.

### **General Equation:**
For a circle with center $(h, k)$ and radius $r$:
$$
(x - h)^2 + (y - k)^2 = r^2
$$

### **Expanded Equation:**
Expanding the general form:
$$
x^2 + y^2 - 2hx - 2ky + h^2 + k^2 - r^2 = 0
$$

### **Key Properties:**
- **Center:** $(h, k)$
- **Radius:** $r$
- **Intercepts:**
  - If the center is at $(0, 0)$, $r$ is the distance to any point on the circle.
  - Solve for $x$ and $y$ intercepts by substituting $y = 0$ and $x = 0$ into the equation.

### **Example:**
Find the equation of a circle with center at $(1, 2)$ and radius $3$:
$$
(x - 1)^2 + (y - 2)^2 = 3^2
$$
Simplify:
$$
(x - 1)^2 + (y - 2)^2 = 9
$$

---

## **2. Parabola**

### **Definition:**
A parabola is the set of all points in a plane equidistant from a fixed point (focus) and a fixed line (directrix).

### **Standard Forms:**
1. **Vertical Parabola:**
   $$
   y = a(x - h)^2 + k
   $$
   - Focus lies on the axis of symmetry above or below the vertex.
   - Opens up if $a > 0$, down if $a < 0$.

2. **Horizontal Parabola:**
   $$
   x = a(y - k)^2 + h
   $$
   - Focus lies to the left or right of the vertex.
   - Opens right if $a > 0$, left if $a < 0$.

### **Key Properties:**
- **Vertex:** $(h, k)$ (minimum or maximum point).
- **Focus:** Located $1/(4a)$ units from the vertex along the axis of symmetry.
- **Directrix:** Line perpendicular to the axis of symmetry, opposite to the focus.

### **Example:**
Find the equation of a parabola passing through $(3, 1)$, intersecting the $x$-axis at $x = 2$ and $x = 4$:
1. The roots are $x = 2$ and $x = 4$, so the equation is:
   $$
   y = a(x - 2)(x - 4)
   $$
2. Use point $(3, 1)$ to find $a$:
   $$
   1 = a(3 - 2)(3 - 4) \implies 1 = a(-1) \implies a = -1
   $$
   Final equation:
   $$
   y = -(x - 2)(x - 4) = -(x^2 - 6x + 8) \implies y = -x^2 + 6x - 8
   $$

---

## **3. Ellipse**

### **Definition:**
An ellipse is the set of all points where the sum of the distances to two fixed points (foci) is constant.

### **Standard Forms:**
1. **Horizontal Major Axis:**
   $$
   \frac{(x - h)^2}{a^2} + \frac{(y - k)^2}{b^2} = 1
   $$
   - Major axis: Horizontal, length $2a$.
   - Minor axis: Vertical, length $2b$.

2. **Vertical Major Axis:**
   $$
   \frac{(x - h)^2}{b^2} + \frac{(y - k)^2}{a^2} = 1
   $$
   - Major axis: Vertical, length $2a$.
   - Minor axis: Horizontal, length $2b$.

### **Key Properties:**
- **Center:** $(h, k)$
- **Foci:** Located at a distance $c$ from the center along the major axis, where:
  $$
  c = \sqrt{a^2 - b^2}
  $$
- **Vertices:** Endpoints of the major axis.

### **Example:**
Find the center of the ellipse:
$$
x^2 + 4y^2 - 4x - 16y + 16 = 0
$$
1. Group and complete the square:
   $$
   (x^2 - 4x) + 4(y^2 - 4y) = -16
   $$
   Completing the square:
   $$
   (x - 2)^2 + 4(y - 2)^2 = 4
   $$
   Simplify:
   $$
   \frac{(x - 2)^2}{4} + \frac{(y - 2)^2}{1} = 1
   $$
   Center: $(2, 2)$

---

## **4. Hyperbola**

### **Definition:**
A hyperbola is the set of all points where the absolute difference of the distances to two fixed points (foci) is constant.

### **Standard Forms:**
1. **Horizontal Transverse Axis:**
   $$
   \frac{(x - h)^2}{a^2} - \frac{(y - k)^2}{b^2} = 1
   $$
   - Transverse axis: Horizontal, length $2a$.
   - Conjugate axis: Vertical, length $2b$.

2. **Vertical Transverse Axis:**
   $$
   \frac{(y - k)^2}{a^2} - \frac{(x - h)^2}{b^2} = 1
   $$
   - Transverse axis: Vertical, length $2a$.
   - Conjugate axis: Horizontal, length $2b$.

### **Key Properties:**
- **Foci:** Located at a distance $c$ from the center along the transverse axis:
  $$
  c = \sqrt{a^2 + b^2}
  $$
- **Asymptotes:**
  - Horizontal: $y - k = \pm \frac{b}{a}(x - h)$
  - Vertical: $y - k = \pm \frac{a}{b}(x - h)$

---

## **5. Tangents to Conic Sections**

### **Circle Tangents:**
For a circle with center $(h, k)$ and radius $r$, a line $y = mx + c$ is tangent if:
$$
|c| = \sqrt{r^2(1 + m^2)}
$$

### **Example:**
Find the slope $m > 0$ of a tangent to the circle:
$$
x^2 + y^2 = 1
$$
Line: $y = mx - 5$. Substitute into the circle equation:
$$
x^2 + (mx - 5)^2 = 1
$$
Solve for $m$ to ensure a single intersection point.

---

## **6. Intersection of Conic Sections**

### Example:
Find the intersection points of:
$$
x^2 - y^2 = 1 \quad \text{and} \quad x^2 + 4y^2 = 6
$$
1. Substitute $x^2 = y^2 + 1$ from the first equation into the second:
   $$
   (y^2 + 1) + 4y^2 = 6
   $$
2. Solve for $y$:
   $$
   5y^2 = 5 \implies y = \pm 1
   $$
3. Solve for $x$:
   $$
   x^2 = 1 + y^2 = 2 \implies x = \pm \sqrt{2}
   $$
Intersection points: $(\sqrt{2}, 1), (-\sqrt{2}, -1)$.

---

## **7. Distance Between Branches of a Hyperbola**

### Horizontal Hyperbola:
For a hyperbola:
$$
\frac{x^2}{a^2} - \frac{y^2}{b^2} = 1
$$
The distance between the branches is:
$$
d = 2b
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a grid for plotting
figure, axes = plt.subplots(3, 2, figsize=(12, 12))
x_vals = np.linspace(-5, 5, 400)
y_vals = np.linspace(-5, 5, 400)
grid_x, grid_y = np.meshgrid(x_vals, y_vals)

# Task 1: Circle
axes[0, 0].set_title("Circle: $(x - 1)^2 + (y - 2)^2 = 9$")
axes[0, 0].contour(grid_x, grid_y, (grid_x - 1)**2 + (grid_y - 2)**2 - 9, levels=[0], colors='blue')
axes[0, 0].grid()

# Task 2: Parabola
x_range = np.linspace(1, 5, 100)
parabola_y = -((x_range - 2) * (x_range - 4))
axes[0, 1].set_title("Parabola: $y = -(x - 2)(x - 4)$")
axes[0, 1].plot(x_range, parabola_y, label="$y = -x^2 + 6x - 8$", color='green')
axes[0, 1].grid()

# Task 3: Ellipse
axes[1, 0].set_title("Ellipse: $(x - 2)^2/4 + (y - 2)^2 = 1$")
axes[1, 0].contour(grid_x, grid_y, (grid_x - 2)**2 / 4 + (grid_y - 2)**2 - 1, levels=[0], colors='red')
axes[1, 0].grid()

# Task 4: Tangent to Circle
slope_m = 2 * np.sqrt(6)
x_line = np.linspace(-1.5, 1.5, 100)
y_line = slope_m * x_line - 5
axes[1, 1].set_title("Circle Tangent: $y = mx - 5$")
axes[1, 1].contour(grid_x, grid_y, grid_x**2 + grid_y**2 - 1, levels=[0], colors='blue')
axes[1, 1].plot(x_line, y_line, label=f"$m = {slope_m:.2f}$", color='orange')
axes[1, 1].grid()

# Task 5: Intersection of Hyperbola and Ellipse
axes[2, 0].set_title("Intersection of Hyperbola and Ellipse")
axes[2, 0].contour(grid_x, grid_y, grid_x**2 - grid_y**2 - 1, levels=[0], colors='blue', label="Hyperbola")
axes[2, 0].contour(grid_x, grid_y, grid_x**2 + 4 * grid_y**2 - 6, levels=[0], colors='red', label="Ellipse")
axes[2, 0].grid()

# Task 6: Hyperbola Branch Distance
axes[2, 1].set_title("Hyperbola: Distance Between Branches")
axes[2, 1].contour(grid_x, grid_y, grid_x**2 - grid_y**2 - 1, levels=[0], colors='blue')
axes[2, 1].axhline(y=1, color='green', linestyle='--', label="Distance = 2")
axes[2, 1].axhline(y=-1, color='green', linestyle='--')
axes[2, 1].grid()

# General plot settings
for subplot in axes.flat:
    subplot.set_xlim([-5, 5])
    subplot.set_ylim([-5, 5])
    subplot.axhline(0, color='black', linewidth=0.5)
    subplot.axvline(0, color='black', linewidth=0.5)

plt.tight_layout()
plt.show()
```
