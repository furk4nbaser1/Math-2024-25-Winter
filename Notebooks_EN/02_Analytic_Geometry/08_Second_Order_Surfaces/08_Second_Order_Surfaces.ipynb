## **1. Write the equation of a sphere with center at point $P=(1, 2, 3)$ and radius $r=3$.**

### **Concept:**
A sphere in 3D space is defined as the set of all points equidistant from a fixed point (the center). The general equation of a sphere is:
$$
(x - h)^2 + (y - k)^2 + (z - l)^2 = r^2
$$
where $(h, k, l)$ is the center of the sphere, and $r$ is the radius.

### **Solution:**
Substituting $P=(1, 2, 3)$ and $r=3$:
$$
(x - 1)^2 + (y - 2)^2 + (z - 3)^2 = 3^2
$$
Simplify:
$$
(x - 1)^2 + (y - 2)^2 + (z - 3)^2 = 9
$$

---

## **2. Do the spheres with equations $x^2 + y^2 + z^2 = 1$ and $x^2 + y^2 + z^2 = 2$ have any common points?**

### **Concept:**
Two spheres intersect if the distance between their centers is less than or equal to the sum of their radii and greater than or equal to the absolute difference of their radii.

1. **Given equations:**
   - Sphere 1: $x^2 + y^2 + z^2 = 1$ (center: $(0, 0, 0)$, radius: $r_1 = \sqrt{1} = 1$)
   - Sphere 2: $x^2 + y^2 + z^2 = 2$ (center: $(0, 0, 0)$, radius: $r_2 = \sqrt{2}$)

2. **Distance between centers:**
   The centers are the same point $(0, 0, 0)$, so the distance is $0$.

3. **Check intersection:**
   Since the distance is $0$, the spheres are concentric. They do not intersect because the radii differ.

### **Conclusion:**
The spheres have no common points.

---

## **3. What curve in space is formed by the intersection of the sphere $x^2 + y^2 + z^2 = 1$ with the sphere $(x-1)^2 + y^2 + z^2 = 1$? Find the equation of this curve.**

### **Concept:**
The intersection of two spheres forms a circle in 3D space.

1. **Given equations:**
   - Sphere 1: $x^2 + y^2 + z^2 = 1$
   - Sphere 2: $(x - 1)^2 + y^2 + z^2 = 1$

2. **Expand Sphere 2:**
   $$
   (x - 1)^2 + y^2 + z^2 = 1 \implies x^2 - 2x + 1 + y^2 + z^2 = 1
   $$
   Simplify:
   $$
   x^2 + y^2 + z^2 - 2x = 0
   $$

3. **Subtract Sphere 1:**
   Subtract $x^2 + y^2 + z^2 = 1$ from $x^2 + y^2 + z^2 - 2x = 0$:
   $$
   -2x = -1 \implies x = \frac{1}{2}
   $$

4. **Substitute into Sphere 1:**
   Substitute $x = \frac{1}{2}$ into $x^2 + y^2 + z^2 = 1$:
   $$
   \left(\frac{1}{2}\right)^2 + y^2 + z^2 = 1
   $$
   Simplify:
   $$
   y^2 + z^2 = \frac{3}{4}
   $$

### **Equation of the Curve:**
The intersection is a circle in the plane $x = \frac{1}{2}$ with radius $\sqrt{\frac{3}{4}} = \frac{\sqrt{3}}{2}$:
$$
y^2 + z^2 = \frac{3}{4}, \quad x = \frac{1}{2}
$$

## **4. Write the equation of the tangent plane to the paraboloid $z=(x-1)^2+y^2+1$ at point $P=(1, 0, 1)$.**

### **Concept:**
The tangent plane to a surface $z = f(x, y)$ at a point $(x_0, y_0, z_0)$ is given by:
$$
z - z_0 = f_x(x_0, y_0)(x - x_0) + f_y(x_0, y_0)(y - y_0)
$$
where $f_x$ and $f_y$ are partial derivatives of $f$ with respect to $x$ and $y$.

1. **Given surface:**
   $$
   z = (x - 1)^2 + y^2 + 1
   $$

2. **Partial derivatives:**
   $$
   f_x = \frac{\partial}{\partial x}[(x - 1)^2 + y^2 + 1] = 2(x - 1)
   $$
   $$
   f_y = \frac{\partial}{\partial y}[(x - 1)^2 + y^2 + 1] = 2y
   $$

3. **Evaluate at $P(1, 0, 1)$:**
   $$
   f_x(1, 0) = 2(1 - 1) = 0, \quad f_y(1, 0) = 2(0) = 0
   $$

4. **Tangent plane equation:**
   $$
   z - 1 = 0(x - 1) + 0(y - 0)
   $$
   Simplify:
   $$
   z = 1
   $$

**Equation:**  
The tangent plane is:
$$
z = 1
$$

```python
import numpy as np
import matplotlib.pyplot as plt

u, v = np.mgrid[0:2*np.pi:100j, 0:np.pi:50j]
x_sphere1 = 1 + 3 * np.cos(u) * np.sin(v)
y_sphere1 = 2 + 3 * np.sin(u) * np.sin(v)
z_sphere1 = 3 + 3 * np.cos(v)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x_sphere1, y_sphere1, z_sphere1, color='blue', alpha=0.5)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title("Sphere with center (1, 2, 3) and radius 3")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

u, v = np.mgrid[0:2*np.pi:100j, 0:np.pi:50j]
x_sphere2 = np.sin(v) * np.cos(u)
y_sphere2 = np.sin(v) * np.sin(u)
z_sphere2 = np.cos(v)
x_sphere3 = np.sqrt(2) * np.sin(v) * np.cos(u)
y_sphere3 = np.sqrt(2) * np.sin(v) * np.sin(u)
z_sphere3 = np.sqrt(2) * np.cos(v)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x_sphere2, y_sphere2, z_sphere2, color='red', alpha=0.3)
ax.plot_surface(x_sphere3, y_sphere3, z_sphere3, color='green', alpha=0.3)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title("Two Spheres with Different Radii")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 2 * np.pi, 100)
circle_y = np.sqrt(3) / 2 * np.cos(theta)
circle_z = np.sqrt(3) / 2 * np.sin(theta)
circle_x = np.full_like(circle_y, 0.5)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(circle_x, circle_y, circle_z, color='purple')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title("Intersection of Two Spheres")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = np.ones_like(X)  # Tangent plane z = 1
Z_paraboloid = (X - 1)**2 + Y**2 + 1

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.5, color='orange')
ax.plot_surface(X, Y, Z_paraboloid, alpha=0.3, color='cyan')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
plt.title("Tangent Plane to Paraboloid at P(1, 0, 1)")
plt.show()
```
