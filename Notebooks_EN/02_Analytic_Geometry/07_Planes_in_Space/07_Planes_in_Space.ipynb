# Equations of Planes in Space: Concepts and Solutions

---

## **1. The plane passes through points $A(1, 2, 3)$, $B(3, 4, 5)$, and $C(2, 1, 4)$.**

### **Concept:**
The equation of a plane passing through three points can be found using the cross product of two vectors formed by these points.  
The general equation of a plane is:
$$
ax + by + cz + d = 0
$$
Where $(a, b, c)$ is the normal vector of the plane.

1. **Find two vectors in the plane:**
   $$ \vec{AB} = B - A = (3 - 1, 4 - 2, 5 - 3) = (2, 2, 2) $$
   $$ \vec{AC} = C - A = (2 - 1, 1 - 2, 4 - 3) = (1, -1, 1) $$

2. **Find the normal vector:**
   $$ \vec{n} = \vec{AB} \times \vec{AC} $$
   $$ \vec{n} = \begin{vmatrix}
   \mathbf{i} & \mathbf{j} & \mathbf{k} \\
   2 & 2 & 2 \\
   1 & -1 & 1
   \end{vmatrix} = \mathbf{i}(2 - (-2)) - \mathbf{j}(2 - 2) + \mathbf{k}(-2 - 2) $$
   $$ \vec{n} = (4, 0, -4) $$

3. **Substitute into the plane equation:**
   $$ 4(x - 1) + 0(y - 2) - 4(z - 3) = 0 $$
   $$ 4x - 4z - 4 = 0 \implies x - z = 1 $$

**Equation:**  
$$ x - z = 1 $$

---

## **2. The plane passes through point $A(1, 2, 3)$ and is parallel to the plane $2x + 3y + 4z = 5$.**

### **Concept:**
Parallel planes have the same normal vector.  
The equation of the plane is:
$$
2x + 3y + 4z + d = 0
$$

1. Substitute point $A(1, 2, 3)$:
   $$ 2(1) + 3(2) + 4(3) + d = 0 $$
   $$ 2 + 6 + 12 + d = 0 \implies d = -20 $$

**Equation:**  
$$ 2x + 3y + 4z - 20 = 0 $$

---

## **3. The plane passes through point $A(1, 2, 3)$ and is perpendicular to the normal vector $\vec{n} = [2, 3, 4]$.**

### **Concept:**
The equation of the plane is:
$$
2x + 3y + 4z + d = 0
$$

1. Substitute point $A(1, 2, 3)$:
   $$ 2(1) + 3(2) + 4(3) + d = 0 $$
   $$ 2 + 6 + 12 + d = 0 \implies d = -20 $$

**Equation:**  
$$ 2x + 3y + 4z - 20 = 0 $$

## **4. Find the line of intersection of planes $2x + 3y + 4z = 5$ and $3x + 4y + 2z = 6$.**

### **Concept:**
The line of intersection is parallel to the cross product of the normals of the planes.

1. **Normals:**  
   $$ \vec{n_1} = [2, 3, 4], \, \vec{n_2} = [3, 4, 2] $$
   $$ \vec{d} = \vec{n_1} \times \vec{n_2} = \begin{vmatrix}
   \mathbf{i} & \mathbf{j} & \mathbf{k} \\
   2 & 3 & 4 \\
   3 & 4 & 2
   \end{vmatrix} = \mathbf{i}(-10 - 16) - \mathbf{j}(4 - 12) + \mathbf{k}(8 - 9) $$
   $$ \vec{d} = [-26, 8, -1] $$

2. **Find a point on the line:**  
   Solve:
   $$ 2x + 3y + 4z = 5 $$
   $$ 3x + 4y + 2z = 6 $$
   Set $z = 0$, solve:
   $$ 2x + 3y = 5, \, 3x + 4y = 6 $$
   $$ x = 2, \, y = -1 $$

**Parametric Equation:**  
$$ x = 2 - 26t, \, y = -1 + 8t, \, z = -t $$

## **5. Plane passing through $A(1, 2, 3)$ parallel to vectors $\vec{v_1} = [1, 0, 1]$ and $\vec{v_2} = [0, 1, -1]$.**

### **Concept:**
The normal vector is:
$$
\vec{n} = \vec{v_1} \times \vec{v_2} = \begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
1 & 0 & 1 \\
0 & 1 & -1
\end{vmatrix} = \mathbf{i}(-1 - 1) - \mathbf{j}(1 - 0) + \mathbf{k}(1 - 0)
$$
$$
\vec{n} = [-2, -1, 1]
$$

1. Substitute into the plane equation:
   $$ -2(x - 1) - (y - 2) + (z - 3) = 0 $$
   $$ -2x - y + z + 5 = 0 $$

**Equation:**  
$$ -2x - y + z + 5 = 0 $$

## **7. Distance from point $A(1, 2, 3)$ to plane $2x + 3y + 4z = 5$.**

### **Concept:**
Distance formula:
$$
d = \frac{|ax_1 + by_1 + cz_1 - d|}{\sqrt{a^2 + b^2 + c^2}}
$$
Substitute $a = 2$, $b = 3$, $c = 4$, $d = 5$, and $A(1, 2, 3)$:
$$
d = \frac{|2(1) + 3(2) + 4(3) - 5|}{\sqrt{2^2 + 3^2 + 4^2}}
$$
$$
d = \frac{|2 + 6 + 12 - 5|}{\sqrt{4 + 9 + 16}} = \frac{15}{\sqrt{29}}
$$

## **9. Angle between planes $x + y + z = 1$ and $x = 0$.**

### **Concept:**
Angle formula:
$$
\cos \theta = \frac{\vec{n_1} \cdot \vec{n_2}}{\|\vec{n_1}\| \|\vec{n_2}\|}
$$
Normals:
$$
\vec{n_1} = [1, 1, 1], \, \vec{n_2} = [1, 0, 0]
$$
$$
\cos \theta = \frac{1}{\sqrt{3} \cdot 1} = \frac{1}{\sqrt{3}}
$$

```python
import numpy as np

# Define points
A = np.array([1, 2, 3])
B = np.array([3, 4, 5])
C = np.array([2, 1, 4])

# Vector AB and AC
AB = B - A
AC = C - A

# Normal vector to the plane through A, B, C (cross product of AB and AC)
normal_vector = np.cross(AB, AC)
d = np.dot(normal_vector, A)

# Equation of the plane through points A, B, C
plane_equation = f"{normal_vector[0]}x + {normal_vector[1]}y + {normal_vector[2]}z = {d}"
print(plane_equation)
```

```python
# Using the same normal vector and point A to find the plane equation
d_parallel = np.dot([2, 3, 4], A)
plane_equation_parallel = f"2x + 3y + 4z = {d_parallel}"
print(plane_equation_parallel)
```

```python
# Since the plane is perpendicular to [2, 3, 4], it has the same normal vector
plane_equation_perpendicular = f"2x + 3y + 4z = {d_parallel}"
print(plane_equation_perpendicular)
```

```python
from sympy import symbols, Eq, solve

x, y, z = symbols('x y z')
plane1 = Eq(2*x + 3*y + 4*z, 5)
plane2 = Eq(3*x + 4*y + 2*z, 6)
line_of_intersection = solve((plane1, plane2), (x, y, z))
print(line_of_intersection)
```

```python
v1 = np.array([1, 0, 1])
v2 = np.array([0, 1, -1])
normal_vector_v1_v2 = np.cross(v1, v2)
d_v1_v2 = np.dot(normal_vector_v1_v2, A)
plane_equation_v1_v2 = f"{normal_vector_v1_v2[0]}x + {normal_vector_v1_v2[1]}y + {normal_vector_v1_v2[2]}z = {d_v1_v2}"
print(plane_equation_v1_v2)
```

```python
cos_angle = np.dot([1, 1, 1], [1, 0, 0]) / (np.linalg.norm([1, 1, 1]) * np.linalg.norm([1, 0, 0]))
angle_between_planes = np.arccos(cos_angle)
print(f"Angle between the planes is {np.degrees(angle_between_planes):.2f} degrees")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Points
A = np.array([1, 2, 3])
B = np.array([3, 4, 5])
C = np.array([2, 1, 4])

# Plane
points = np.array([A, B, C])
x_range = np.linspace(0, 5, 10)
y_range = np.linspace(0, 5, 10)
X, Y = np.meshgrid(x_range, y_range)
Z = (2 - 2*X + Y)/2  # From cross product and point

# Plot points
ax.scatter(points[:, 0], points[:, 1], points[:, 2], color='r')

# Plot plane
ax.plot_surface(X, Y, Z, alpha=0.5, rstride=100, cstride=100, color='b', edgecolor='k')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('Plane through A, B, C')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Using the point A and the normal of the given plane
d = 2*A[0] + 3*A[1] + 4*A[2]  # Calculate d using point A
Z = (d - 2*X - 3*Y) / 4

# Plot point A
ax.scatter(A[0], A[1], A[2], color='r')

# Plot parallel plane
ax.plot_surface(X, Y, Z, alpha=0.5, rstride=100, cstride=100, color='g', edgecolor='k')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('Plane through A parallel to given plane')
plt.show()
```
